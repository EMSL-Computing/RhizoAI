# Preprocess and flatten image

```python
import numpy as np
from astropy.io import fits

import cv2 as cv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import copy
import os
import math
```

```python
#image_path = r'./volumes_fits_24Mar2024/'
cwd = os.getcwd()
image_path_mar = os.path.abspath(os.path.join(cwd, os.pardir)) + "/OCT_data/volumes_FITS_29Mar2024/"

file_paths_mar = [image_path_mar + x for x in os.listdir(image_path_mar)]

image_path_may = os.path.abspath(os.path.join(cwd, os.pardir)) + "/OCT_data/volumes_FITS_24May2024/"

file_paths_may = [image_path_may + x for x in os.listdir(image_path_may)]

all_file_paths = file_paths_mar + file_paths_may

count_decay = 0
count_dead = 0
count_live = 0

index = 0

for file_path in all_file_paths:
    ### Categorizing and counting the data. We take 3 slices from each FITS, so we increment the count +3
    print(file_path)
    if 'Decaying' in file_path:
        category = 'decay'
        count_decay += 3
    elif 'Dead' in file_path:
        category = 'dead'
        count_dead += 3
    else:
        category = 'live'
        count_live += 3

    export_name = "mar+may_2024_threshold/" + str(index) + "_" + category


    raw = fits.getdata(file_path)
    half = math.floor(raw.shape[0] / 2)
    img_f = copy.deepcopy(raw[0,:,:])
    img_m = copy.deepcopy(raw[half,:,:])
    img_l = copy.deepcopy(raw[raw.shape[0]-1,:,:])

    ret, img_first = cv.threshold(img_f, 127, 255, cv.THRESH_BINARY)
    ret, img_middle = cv.threshold(img_m, 127, 255, cv.THRESH_BINARY)
    ret, img_last = cv.threshold(img_l, 127, 255, cv.THRESH_BINARY)

    if not os.path.isfile(export_name + "0.npy"):
        np.save(export_name + "0", img_first)
    if not os.path.isfile(export_name + str(half) + '.npy'):
        np.save(export_name + str(half), img_middle)
    if not os.path.isfile(export_name + str(raw.shape[0]-1) + '.npy'):
        np.save(export_name + str(raw.shape[0]-1), img_last)

    category = ""


    #cv.imwrite("29mar2024_data_image_raw/" + filename[:3] + "_" + category + "0.png", img_f)
    #cv.imwrite("29mar2024_data_image_raw/" + filename[:3] + "_" + category + str(half) + ".png", img_m)
    #cv.imwrite("29mar2024_data_image_raw/" + filename[:3] + "_" + category + str(raw.shape[0]-1) + ".png", img_l)

    #np.save("mar+may_2024_raw/" + str(index) + "_" + category + "0", img_f)
    #np.save("mar+may_2024_raw/" + str(index) + "_" + category + str(half), img_m)
    #np.save("mar+may_2024_raw/" + str(index) + "_" + category + str(raw.shape[0]-1), img_l)

    index += 1
```
